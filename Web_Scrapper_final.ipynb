```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
if __name__ == '__main__':
    title = []
    Price = []
    product_link_list = []
    Rating = []
    decr = []
    
    for i in range(1,11):
        url = 'https://www.flipkart.com/search?q=5g+smartphone+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+ str(i)
        HEADERS =({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36','Accept-Language':'en-US, en;q=0.5'})
        webpage = requests.get(url, headers=HEADERS)
        soup = BeautifulSoup(webpage.content, "lxml")
        box = soup.find('div', class_= '_1YokD2 _3Mn1Gg')
        
        titles = box.find_all('div', class_ = '_4rR01T')
        for i in titles:
            name = i.text
            title.append(name)
        
        prices = box.find_all('div', class_ = '_30jeq3 _1_WHN1')
        for i in prices:
            name = i.text
            Price.append(name)
        
        links = box.find_all('a', class_ = '_1fQZEK')
        for link in links:
            product = "https://www.flipkart.com"+ link.get("href")
            product_link_list.append(product)
        
        rating = box.find_all('div', class_ = '_3LWZlK')
        for i in rating:
            name = i.text
            Rating.append(name)

        description = box.find_all('ul', class_ = '_1xgFaf')
        for i in description:
            name = i.text
            decr.append(name)
        
    df = pd.DataFrame({'Product_name': title, 'Product_Link': product_link_list,'Price': Price, 'Rating': Rating, 'Description': decr})
```

```python
df.to_csv("flipkart_phones_under_50000.csv", header=True, index=False)
```
